```python
import pandas as pd
import numpy as np
import json
from IPython.display import display, HTML

vg_dataframe = pd.read_csv("data/vgsales.csv")

# Just check on the dataset
vg_dataframe.head()
```

```python
nintendo_dataframe = vg_dataframe.loc[vg_dataframe['Publisher'] == "Nintendo"]
nintendo_dataframe["Platform"].unique()
```

```python
dataframes = []

#Grab top 5 wii games
wii_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "Wii"][:5]
dataframes.append(wii_dataframe)

#Grab top 5 nes games
nes_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "NES"][:5]
dataframes.append(nes_dataframe)

#Grab top 5 DS games
ds_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "DS"][:5]
dataframes.append(ds_dataframe)

#Grab top 5 3DS games
threeDS_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "3DS"][:5]
dataframes.append(threeDS_dataframe)

#Grab top 5 GB games
gb_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "GB"][:5]
dataframes.append(gb_dataframe)

#Grab top 5 GBA games
gba_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "GBA"][:5]
dataframes.append(gba_dataframe)

#Grab top 5 SNES games
snes_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "SNES"][:5]
dataframes.append(snes_dataframe)

#Grab top 5 N64 games
n64_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "N64"][:5]
dataframes.append(n64_dataframe)

#Grab top 5 GC games
gc_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "GC"][:5]
dataframes.append(gc_dataframe)

#Grab top 5 Wiiu games
wiiu_dataframe = vg_dataframe.loc[vg_dataframe['Platform'] == "WiiU"][:5]
dataframes.append(wiiu_dataframe)
```

```python
# Looking at how we may want to convert to json
wii_dataframe.to_json("data/wii.json")
nes_dataframe.to_json("data/nes.json")
ds_dataframe.to_json("data/ds.json")
threeDS_dataframe.to_json("data/threeDS.json")
gb_dataframe.to_json("data/gb.json")
gba_dataframe.to_json("data/gba.json")
snes_dataframe.to_json("data/snes.json")
n64_dataframe.to_json("data/n64.json")
gc_dataframe.to_json("data/gc.json")
wiiu_dataframe.to_json("data/wiiu.json")

# for gif finding
# print(wii_dataframe["Name"])
# print(nes_dataframe["Name"])
# print(ds_dataframe["Name"])
# print(threeDS_dataframe["Name"])
# print(gb_dataframe["Name"])
# print(gba_dataframe["Name"])
# print(snes_dataframe["Name"])
# print(n64_dataframe["Name"])
# print(gc_dataframe["Name"])
# print(wiiu_dataframe["Name"])

output_json = {
    "Wii" : [],
    "NES" : [],
    "DS" : [],
    "threeDS" : [],
    "GB" : [],
    "GBA" : [],
    "SNES" : [],
    "N64" : [],
    "GC" : [],
    "WiiU" : [],
}

result = pd.concat(dataframes)

# Loop through the dataframe, creating json objects and adding them to list
for index, row in result.iterrows():
#     print(row.to_dict())
    if row["Platform"] == "3DS":
        # we need a special case for 3DS as we don't want a variable name starting with a number
        output_json["threeDS"].append(row.to_dict())
    else:
        output_json[row["Platform"]].append(row.to_dict())
    
with open('data/nintendo.json', 'w') as outfile:
    json.dump(output_json, outfile, indent=4)
    
```
